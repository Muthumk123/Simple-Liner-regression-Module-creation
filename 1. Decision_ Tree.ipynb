```python
import pandas as pd
```

```python
dataset =pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent
```

```python
depentent=dataset[['Profit']]
depentent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,depentent,test_size=0.30,random_state=0) # spliting the input and output
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion="poisson",splitter='best') # This turning parameter
regressor=regressor.fit(X_train,y_train)  # model creation
```

```python
import matplotlib.pyplot as plt
from sklearn import tree # seeing the tree like graph
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
