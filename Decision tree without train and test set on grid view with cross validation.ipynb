```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
indep
```

```python
dep=dataset[['charges']]
dep
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
pream_grid={'splitter':['best','random'],'criterion':['squared_error','absolute_error','poisson'] ,'random_state':[0]}
grid=GridSearchCV(DecisionTreeRegressor(),pream_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
re=grid.cv_results_
print("The R_score value of the best perameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])#change the paramter,play with it.
print("Future_Prediction=",Future_Prediction) # also we can call print("Future_Prediction={}".format(Future_Prediction))
```
