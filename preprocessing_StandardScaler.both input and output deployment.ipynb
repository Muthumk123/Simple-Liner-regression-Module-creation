```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
depentent=dataset[['Profit']]
X_train,X_test,y_train,y_test=train_test_split(independent,depentent,test_size=0.30,random_state=0)

sc=StandardScaler()
X_train=sc.fit_transform(X_train) # standardtion for input to work better block
X_test=sc.transform(X_test)

scy=StandardScaler()
y_train=scy.fit_transform(y_train) # standardtion for output to work better block
y_test=scy.transform(y_test)
```
